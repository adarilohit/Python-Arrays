```python
#Exploratory Data Analysis (EDA) is a important step in data science as it visualizing data to understand its main features, find patterns and discover how different parts of the data are connected.
```

```python
#Univariate Analysis: focuses on studying one variable to understand its characteristics
```

```python
import pandas as pd
import seaborn as sns
data = pd.read_csv("C:/Users/adari/Downloads/Employee_dataset.csv")
print(data.head())
```

```python
#univariate analysis on Numerical variables using the histogram function.
sns.histplot(data['age'])
```

```python
#Univariate analysis of categorical data
sns.countplot(data['gender_full'])
```

```python
#A piechart helps us to visualize the percentage of the data belonging to each category.
import matplotlib.pyplot as plt

x = data['STATUS_YEAR'].value_counts()
plt.pie(x.values,
        labels=x.index,
        autopct='%1.1f%%')   #is a string used to label the wedge with their numerical value. 
plt.show()
```

```python
#Bivariate analysis is the simultaneous analysis of two variables. 
# Categorical v/s Numerical
# Numerical V/s Numerical
# Categorical V/s Categorical data
```

```python
#Categorical v/s Numerical
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
sns.barplot(x=data['department_name'], y=data['length_of_service'])
plt.xticks(rotation='vertical')
```

```python
#Black horizontal line is indicating huge differences in the length of service among different departments.
```

```python
#Numerical v/s Numerical
sns.scatterplot(x=data['length_of_service'],
                y=data['age'])
```

```python
#Categorical v/s Categorical
sns.countplot(x='STATUS_YEAR', hue='STATUS', data=data)
```

```python
#Multivariate Analysis
#it involves multiple variables at the same time to find correlation between them.
```

```python
from sklearn import datasets, decomposition
iris = datasets.load_iris()
X = iris.data
y = iris.target
pca = decomposition.PCA(n_components=2)
X = pca.fit_transform(X)
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Only use numeric columns for correlation
numeric_data = data.select_dtypes(include=['number'])

# Plot the heatmap
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```
